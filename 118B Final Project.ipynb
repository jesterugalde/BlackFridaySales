```python
#Using Pandas to Clean Data

import pandas as pd
import numpy as np

df = pd.read_csv('BlackFriday.csv')

#Fill product category NaN values with 0
df.Product_Category_1 = df.Product_Category_1.fillna('0')
df.Product_Category_2 = df.Product_Category_2.fillna('0')
df.Product_Category_3 = df.Product_Category_3.fillna('0')

#Dropping identifiers, not needed
df.drop('User_ID', inplace = True, axis = 1)

#Dropping '+' sign in stay_in_current_years df
df.Stay_In_Current_City_Years = df.Stay_In_Current_City_Years.apply(lambda x:str(x).replace('+',''))

#Replacing A,B,C categories to 1,2,3 respectively
df.City_Category = df.City_Category.apply(lambda x:str(x).replace('A', '1'))
df.City_Category = df.City_Category.apply(lambda x:str(x).replace('B', '2'))
df.City_Category = df.City_Category.apply(lambda x:str(x).replace('C', '3'))

#Replacing M/F with 0/1 respectively
df.Gender = df.Gender.apply(lambda x:str(x).replace('M', '0'))
df.Gender = df.Gender.apply(lambda x:str(x).replace('F', '1'))

#df.to_csv('BlackFriday1.csv', encoding = 'utf-8')

df.head(10)
```
